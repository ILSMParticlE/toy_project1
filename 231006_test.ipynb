```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

def url_to_text(url):
    webpage = urlopen(url)
    source = BeautifulSoup(webpage, 'lxml')
    texts_html = source.findAll(['p', 'dd', 'dt'])
    texts_list = list(map(lambda x: x.get_text(), texts_html))
    texts = ''
    for text in texts_list:
      if text in texts: continue
      texts += text
    texts = texts.lower()
    return texts

url_list = ["https://web.stanford.edu/class/cs140/projects/pintos/pintos_2.html#SEC17",
            "https://web.stanford.edu/class/cs140/projects/pintos/pintos_3.html#SEC32",
            "https://web.stanford.edu/class/cs140/projects/pintos/pintos_4.html#SEC53",
            "https://web.stanford.edu/class/cs140/projects/pintos/pintos_5.html#SEC75"]

url_list = ["https://web.stanford.edu/class/cs140/projects/pintos/pintos_2.html#SEC17"]

texts = ""
for url in url_list:
    texts += url_to_text(url)

print(texts)
```

###1. 구체화된 토큰

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

texts = texts.lower()
#texts = replacer.replace(texts)
sentence_list = sent_tokenize(texts)


word_list = []
for sentence in sentence_list:
  word_list += word_tokenize(sentence)


tag_list = pos_tag(word_list)
new_word_list = []

# define stop words list
stop_word_list = list(stopwords.words('english'))
stop_word_list += ['.', ',', 'is', 'n\'t']
stop_word_list += ['~', '!', '(', ')', ':', '{', '}', '[', ']']


pos_noun = ['NN', 'NNS', 'NNP', 'NNPS']
pos_verb = ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']

ppw_list = []     # preprocessed words list
ppw = ''


for i in range(len(tag_list)):
  word = tag_list[i][0]
  tag = tag_list[i][1]

  if word in stop_word_list or word[0] == '\'' or tag in ['MD', 'UH']:
    if len(ppw) > 0: ppw_list.append(ppw)
    ppw = ''
  elif tag in pos_verb:
    if len(ppw) > 0: ppw_list.append(ppw)
    ppw_list.append(word)
    ppw = ''
  else:
    if len(ppw) == 0: ppw = word
    else: ppw += ' ' + word

for ppw in ppw_list:
  if ' ' in ppw: print(ppw)



#stemmer = PorterStemmer()
#stemmer = LancasterStemmer()
#stemer = RegexpStemmer("ing")
#preprocessed_word_list = [stemmer.stem(word) for word in preprocessed_word_list]
```

## 출력

```python
word_freq  = {}
for word in ppw_list:
  if word in word_freq.keys():
    word_freq[word] += 1
  else:
    word_freq[word] = 1

word_cnt = []
for word, freq, in word_freq.items():
  word_cnt.append([word, freq])
word_cnt.sort(key=lambda elem: elem[1], reverse=True)

for word, freq in word_cnt:
   print (word + ' ' + str(freq))

#for i in range(40):
#  print (word_cnt[i][0] + ' ' + str(word_cnt[i][1]))
```
